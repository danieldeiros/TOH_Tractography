```python
# Max dose
max_doses_case_1 = [3503.09, 3390.01, 3493.31, 3777.03, 3567.07]
max_doses_case_2 = [3484.00, 3352.54, 3516.72, 3683.43, 3570.93]
max_doses_case_3 = [3498.60, 3496.80, 3816.65, 3673.24, 3584.79]
case_labels = ["3D-CRT (LR-RL)", "3D-CRT (Chosen Gantry Angles)", "IMRT", "VMAT", "CyberKnife"]

plt.figure(figsize=(12,4))
plt.plot(case_labels, max_doses_case_1, 'r.-', markersize = 12, label = "Case 1")
plt.plot(case_labels, max_doses_case_2, 'g.-', markersize = 12, label = "Case 2")
plt.plot(case_labels, max_doses_case_3, 'b.-', markersize = 12, label = "Case 3")

plt.legend()
plt.xlabel("Technique")
plt.ylabel("Max Dose (cGy)")
plt.title("Max Dose for Different Techniques in Different Cases")
```
